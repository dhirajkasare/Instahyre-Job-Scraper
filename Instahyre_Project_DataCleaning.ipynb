```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Instahyre_Project_Scraped_File.csv")
df.head()
```

```python
# checking no_of_rows and columns
df.shape
```

```python
df.dtypes
```

```python
# Clean the skills data by replacing newline characters with commas
df['Skills'] = df['Skills'].str.replace('\n', ', ')

df.head()
```

```python
# Modify the Title column by appending Job Position values
df['Title'] = df['Title'] + ' - ' + df['Job Position']

# Rename the Title column to Company Name
df = df.rename(columns={'Title': 'Company Name'})

# Drop the Job Position column
df = df.drop(columns=['Job Position'])

df.head()
```

```python
# Modify the Location column to include "Job available in [Location]"
df['Location'] = 'Job available in ' + df['Location']

df.head()
```

```python
# checking duplicates rows in DataFrame
print(f"Total Duplicate Rows : {df.duplicated().sum()}")
print()
df[df.duplicated]
```

```python
df.drop_duplicates(inplace=True)
print(f"Total Duplicate Rows : {df.duplicated().sum()}")
```

```python
# checking Null Values in DataFrame
df.isnull().sum()
```

```python
for i in df:
    df[i]=df[i].fillna(df[i].mode()[0])
df.isnull().sum()
```

```python
df.columns
```

```python
for column in df:
    print(f"Column: {column}")
    
    # Print the number of unique values in the column
    unique_count = df[column].nunique()
    print(f"Number of unique values: {unique_count}\n")
    
    if unique_count <= 100:
        print("Unique values:", df[column].unique())
    else:
        print("Too many unique values to display")
        
    print("-" * 100)
```
